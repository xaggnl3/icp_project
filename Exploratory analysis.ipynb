## 1st step : Bring 20 minutes CSV to dataframe we will call df_minutes

```python
import numpy as np
import pandas as pd
df_minutes = pd.read_csv('Articles_Lemonde.csv', index_col=None, header=0, engine='python' )
df_minutes.head()
                         
```

## 2nd step : Add a start date and end date that we can change whenever we want

```python
start_date = '2019-05-01'
end_date = '2019-05-30'
```

```python
mask = (df_minutes['Article_Date'] > start_date) & (df_minutes['Article_Date'] <= end_date)
```

```python
dfa = df_minutes.loc[mask]
dfa
```

## 3rd step: Use the new dataframe with selected dates and use str contains in Article_Title or Article_Summary to obtain data on number of articles for a specific keyword

```python
summarya = dfa.loc[dfa['Article_Title'].str.contains("coronavirus", case=False)]
print (summarya)
```

```python
summarya = dfa.loc[dfa['Article_Summary'].str.contains("coronavirus", case=False)]
print (summarya)
```

```python
summarya = dfa.loc[dfa['Article_Title'].str.contains("vaccination", case=False)]
print (summarya)
```

```python
summarya = dfa.loc[dfa['Article_Summary'].str.contains("vaccination", case=False)]
print (summarya)
```

```python
summarya = dfa.loc[dfa['Article_Title'].str.contains("confinement", case=False)]
print (summarya)
```

```python
summarya = dfa.loc[dfa['Article_Summary'].str.contains("confinement", case=False)]
print (summarya)
```
