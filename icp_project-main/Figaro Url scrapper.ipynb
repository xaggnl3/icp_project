# Le monde

```python
date_range = pd.date_range(start = "2019-09-01",
                           end  ="2021-05-13",
                           freq="D")
```

```python
# dates for beautifulsoup loop
dates = []
for date in date_range:
    date = date.strftime("%d-%m-%Y/")
    dates.append(date)
```

```python
archive_urls = []

for date in dates:
    url = f"https://www.lemonde.fr/archives-du-monde/{date}"
    archive_urls.append(url)
```

# Le Figaro

```python
# tag covid
#latest page
# https://plus.lefigaro.fr/tag/covid-19?page=413
```

```python
# dates for beautifulsoup loop
numbers = []
for num in range(1,414):
    numbers.append(num)
```

```python
figaro_archive_urls = []

for num in numbers:
    url = f"https://plus.lefigaro.fr/tag/covid-19?page={num}"
    figaro_archive_urls.append(url)
```

```python
figaro_archive_urls
```

```python
covid_figaro_urls = []
counter = len(figaro_archive_urls)
for url in figaro_archive_urls:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    tag = soup.find_all("h2", {"itemprop":"name"})
    for url in tag:
        covid_figaro_urls.append(url.a["href"])
    counter -=1
    print(counter)
```

```python
len(covid_figaro_urls)
```

```python
figaro_df = pd.DataFrame(covid_figaro_urls, columns=["url_links"])
```
