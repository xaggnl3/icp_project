```python
import pandas as pd
from selenium import webdriver
import time
import os
import requests
import httplib2
from gazpacho import Soup
from bs4 import BeautifulSoup, SoupStrainer, Tag
from selenium import webdriver
```

# Get date_range for archive search in url bar

```python
date_range = pd.date_range(start = "2019-09-01",
                           end  ="2021-05-13",
                           freq="D")
```

```python
# dates for beautifulsoup loop
dates = []
for date in date_range:
    date = date.strftime("%Y/%m-%d")
    dates.append(date)
```

# Create all archive urls

```python
archive_urls = []

for date in dates:
    url = f"https://www.20minutes.fr/archives/{date}"
    archive_urls.append(url)
```

```python
archive_urls
```

```python
soup_elements = []
counter = len(archive_urls)
for url in archive_urls:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    tag = soup.find("ul", {"class":"spreadlist"})
    soup_elements.append(tag)
    counter -= 1
    print(counter)
```

# Parse lines in elements 

```python
messy_links = []

for element in soup_elements:
    for line in element:
        if isinstance(line, Tag):
            messy_links.append(line)
```

# Append links with href from elements

```python
links = []

for link in messy_links:
    links.append(str(link.a.attrs["href"]))
```

# Join https://www.20minutes.fr if link has key_word

```python
https_links = []

for link in links:
    full_link = "https://www.20minutes.fr" + link
    https_links.append(full_link)
```

```python
len(https_links)
```

```python
key_words = ["sras", "wuhan", "coronavirus", 
             "covid", "covid19",
             "sars-cov-2", "pandemie"]

final_https_links = []

for word in key_words:
    for link in https_links:
        if word in link.lower():
            final_https_links.append(link)
```

```python
link_dataframe = pd.DataFrame(final_https_links, columns=["url_links"])
```

```python
link_dataframe.to_csv("urls_20minutes.csv", sep=",", columns=["url_links"])
```
