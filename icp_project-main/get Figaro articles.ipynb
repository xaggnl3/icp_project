```python
counter = len(figaro_urls["url_links"][5525:])

for url in figaro_urls["url_links"][5525:]:
    
    page = browser.get(url)
    html = browser.page_source
    soup = BeautifulSoup(html, "html.parser")
    
    soup_datetime = soup.find("time")
    date_time = soup_datetime["datetime"]
    
    if "à" in date_time:
        date_time = date_time[:10]
        article_time = pd.to_datetime(date_time).strftime("%Y-%m-%d")
    else:
        article_time = pd.to_datetime(date_time).strftime("%Y-%m-%d")
    
    soup_title = soup.find("p", {"class":"fig-headline-tabbar"})
    if soup_title is None:
        soup_title = soup.find("h1")
        article_title = soup_title.text
    else:
        article_title = soup_title.text
    
    soup_summary = soup.find("p", {"class":"fig-standfirst"})
    if soup_summary is None:
        article_summary = "Pas de Summary pour cet article"
    else:
        article_summary = soup_summary.text.strip()

    soup_article = soup.find_all(["p","h2"], attrs={"class":
                                                    ["fig-paragraph",
                                                     "fig-body-heading"]})
    if soup_article is None:
        soup_article = soup.find_all(["p", "H2"], attrs={"class":["",""]})
        if soup_article is None:
            soup_article = "pas d'article de récuperé. recommencer pour ce lien"
    
    article_frame = []
    
    for paragraph in soup_article:
        if type(paragraph) != str:
            text = paragraph.text.strip()
            article_frame.append(text)

            article = reduce(lambda para_left, para_right: para_left + " " + para_right, article_frame)
            article = article.strip()
            
        else:
            article = "pas d'article de récuperé. recommencer pour ce lien"

    article_url = url

    article_dataframe = article_dataframe.append({
        "Article_Date": article_time, 
        "Article_Title": article_title,
        "Article_Summary": article_summary,
        "Article": article,
        "Article_url": article_url
    }, ignore_index=True)

    print(f"Article n* {counter} appended !")
    counter -= 1
    
```
